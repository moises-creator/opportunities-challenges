# **Prática - Dados climáticos**

### Limpeza e tratamento dos dados


```python
#importando as libs que vamos usar
import pandas as pd
import seaborn as srn
import statistics as sts
```

```python
#lendo o arquivo csv
dataset = pd.read_csv('tempo.csv', sep=';')
dataset
```

```python
#tratar aparencia
agrupado = dataset.groupby(['Aparencia']).size()
agrupado
```

```python
#padronizar de acordo com o dominio
dataset.loc[dataset['Aparencia'] == 'menos', 'Aparencia'] = 'chuva'
```

```python
#visualizar as alteracoes
agrupado = dataset.groupby(['Aparencia']).size()
agrupado
```

```python
#tratar temperatura
dataset['Temperatura'].describe()
```

```python
#visualizar temperatura fora do dominio
dataset.loc[(dataset['Temperatura'] < -130) | (dataset['Temperatura'] > 130)]
```

```python
#calcular a mediana para subsituir o valor fora do dominio
mediana = sts.median(dataset['Temperatura'])
mediana
```

```python
#substituir valor fora do dominio
dataset.loc[(dataset['Temperatura'] < -130) | (dataset['Temperatura'] > 130), 'Temperatura'] = mediana
```

```python
#visualizar novamente para ver se passou alguma coisa
dataset.loc[(dataset['Temperatura'] < -130) | (dataset['Temperatura'] > 130)]
```

```python
#visualizar coluna temperatura tratada
dataset
```

```python
#tratar umidade
dataset['Umidade'].describe()
```

```python
#verificar valores fora do dominio(0-100)
dataset.loc[(dataset['Umidade'] < 0) | (dataset['Umidade'] > 100)]
```

```python
#criar mediana de umidade
mediana = sts.median(dataset['Umidade'])
mediana
```

```python
#subsituir valores fora do dominio com a mediana
dataset.loc[(dataset['Umidade'] < 0) | (dataset['Umidade'] > 100), 'Umidade'] = mediana
```

```python
#substituir valores NAs com a mediana
dataset['Umidade'].fillna(mediana, inplace=True)
```
