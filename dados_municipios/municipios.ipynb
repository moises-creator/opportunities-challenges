# Desafio visualização e análise exploratória de diárias em municípios gaúchos e PIB do município

```python
# Importação das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
#Carregamento da base de dados
dados = pd.read_csv('dados.csv', sep = ';')
dados.head()
```

```python
# Cria um histograma da coluna PIB com kde (densidade de probabilidade) e 10 bins, e adiciona um título ao gráfico
srn.histplot(dados.PIB, kde=True, bins=10).set(title='PIB')
```

```python
# Cria um histograma da coluna VALOREMPENHO com kde (densidade de probabilidade) e 10 bins, e adiciona um título ao gráfico
srn.histplot(dados.VALOREMPENHO, kde=True, bins=10).set(title='Empenho')
```

```python
# Histograma com 10 divisões(bins) e com gráfico de densidade(Kernel density - suaviação)
#stat: count, frequency, probability, percent, density
srn.histplot(dados.PIB, color='blue', kde=True, stat='density', bins=10).set(title='PIB')
```

```python
# Histograma com 10 divisões(bins) e com gráfico de densidade(Kernel density - suaviação)
#stat: count, frequency, probability, percent, density
srn.histplot(dados.VALOREMPENHO, color='blue', kde=True, stat='density', bins=10).set(title='PIB')
```

```python
# Cria um gráfico de dispersão entre as colunas PIB e VALOREMPENHO, usando dados do dataframe, com jitter = False (sem dispersão aleatória)
srn.catplot(x = 'PIB', y = 'VALOREMPENHO', data = dados, jitter = False)
```

```python
# Cria um boxplot dos dados do dataframe sem mostrar os outliers (pontos fora da curva)
srn.boxplot(data = dados, showfliers = False)
```

```python
# Cria um boxplot dos dados do dataframe mostrando os outliers (pontos fora da curva)
srn.boxplot(data = dados, showfliers = True)
```

```python
#visualização das 10 primeiras linhas de dados
dados.head(10)
```

```python
# Cria um gráfico com 6 subplots (2 linhas e 3 colunas) e plota um histograma de PIB para
# cada uma das 6 cidades selecionadas, adicionando um título para cada gráfico

from PIL.PngImagePlugin import o8
#impressão em gráfico 2 x 3 - 3 linhas e 2 colunas
plt.figure()
plt.subplot(3,2,1)
srn.histplot(dados.loc[dados['MUNICIPIO'] == 'SANTANA DO LIVRAMENTO'].PIB, kde=True).set(title='SANTANA DO LIVRAMENTO')
plt.subplot(3,2,2)
srn.histplot(dados.loc[dados['MUNICIPIO'] == 'SANTO ANGELO'].PIB, kde=True).set(title='SANTO ANGELO')
plt.subplot(3,2,3)
srn.histplot(dados.loc[dados['MUNICIPIO'] == 'SAO FRANCISCO DE ASSIS'].PIB, kde=True).set(title='SAO FRANCISCO DE ASSIS')
plt.subplot(3,2,4)
srn.histplot(dados.loc[dados['MUNICIPIO'] == 'SAO GABRIEL'].PIB, kde=True).set(title='SAO GABRIEL')
plt.subplot(3,2,5)
srn.histplot(dados.loc[dados['MUNICIPIO'] == 'QUARAI'].PIB, kde=True).set(title='QUARAI')
plt.subplot(3,2,6)
srn.histplot(dados.loc[dados['MUNICIPIO'] == 'PIRATINI'].PIB, kde=True).set(title='PIRATINI')

#ajusta o layout para não haver sobreposição
plt.tight_layout()
```

```python
# Seleciona as 10 cidades com menor PIB
agrupado = dados.sort_values('PIB').head(10)

# Seleciona apenas as colunas 'MUNICIPIO' e 'PIB' do DataFrame 'agrupado'
agrupado = agrupado.iloc[:,1:3]

# Plota um gráfico de barras com as cidades no eixo x e o PIB no eixo y
agrupado.plot.bar(x='MUNICIPIO',y='PIB', color = 'green')
```

```python
# Seleciona as 10 cidades com menor valor de empenho
agrupado = dados.sort_values('VALOREMPENHO').head(10)

# Seleciona apenas as colunas 'MUNICIPIO' e 'VALOREMPENHO' do DataFrame 'agrupado'
agrupado = agrupado.iloc[:,[1,3]]

# Plota um gráfico de barras com as cidades no eixo x e o valor de empenho no eixo y
agrupado.plot.bar(x='MUNICIPIO',y='VALOREMPENHO', color = 'green')
```
